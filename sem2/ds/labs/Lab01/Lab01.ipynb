```python
# 7a
t = var('t')
x(t) = (1-cos(t))*cos(t)
y(t) = (1-cos(t))*sin(t)
parametric_plot((x(t), y(t)), (t, 0, 2*pi))
```

```python
# 7b
t = var('t')
x(t) = sin(3*t)*cos(t)
y(t) = sin(3*t)*sin(t)
parametric_plot((x(t), y(t)), (t, 0, 6*pi))
```

```python
# 7c
t = var('t')
x(t) = t-sin(t)
y(t) = 1-cos(t)
parametric_plot((x(t), y(t)), (t, 0, 6*pi))
```

```python
# 7d
t = var('t')
x(t) = cos(t) + cos(7*t)/2 + sin(17*t)/3
y(t) = sin(t) + sin(7*t)/2 + cos(17*t)/3
parametric_plot((x(t), y(t)), (t, 0, 2*pi))
```

```python
# 8a
t,s = var('t,s')
f(t, s) = 1-(s*cos(4*t)*cos(t))/sqrt(1-s^2*cos(4*t)^2*sin(t)^2)
x(t, s) = f(t-pi/2, s)
y(t, s) = f(t, s)
parametric_plot((x(t, 0.5), y(t, 0.5)), (t, 0, 2*pi))
```

```python
# 8b
t = var('t')
curves = parametric_plot((x(t, 0.1), y(t, 0.1)), (t, 0, 2*pi))
for i in [2..10]:
    g = parametric_plot((x(t, i/10), y(t, i/10)), (t, 0, 2*pi))
    curves = curves + g
show(curves)
```
